````python
import pandas as pd
import numpy as np

Pessoas_fisicas_data = pd.read_excel('Pessoas_fisicas.xlsx')
Salarios_menores_2k = np.array(Pessoas_fisicas_data["Salario"][Pessoas_fisicas_data["Salario"] >= 2000])
Menor_salario = Salarios_menores_2k.max()

Pessoas_fisicas_data[Pessoas_fisicas_data['Salario'] == Menor_salario]
Nova_coluna = np.random.randn(len(Pessoas_fisicas_data))
Pessoas_fisicas_data["Dados novos"] = Nova_coluna
Pessoas_fisicas_data["Salario Por idade"] = Pessoas_fisicas_data["Salario"]/Pessoas_fisicas_data['Idade']
Pessoas_fisicas_data

from matplotlib import pyplot as plt
Pessoas_fisicas_data.plot(kind='scatter', x='Idade', y='Salario', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
````
